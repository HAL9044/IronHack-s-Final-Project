```python
from bs4 import BeautifulSoup
import time
from datetime import datetime
import pandas as pd
```

```python
def text_cleaner(txt):
    txts = txt.replace("\\u00E1", "á")
    txts = txts.replace("\\u00E9", "é")
    txts = txts.replace("\\u00ED", "í")
    txts = txts.replace("\\u00F3", "ó")
    txts = txts.replace("\\u00FA", "ú")
    txts = txts.replace("\\u00F1", "ñ")
    txts = txts.replace("\\u00C1", "Á")
    txts = txts.replace("\\u00C9", "É")
    txts = txts.replace("\\u00CD", "Í")
    txts = txts.replace("\\u00D3", "Ó")
    txts = txts.replace("\\u00DA", "Ú")
    txts = txts.replace("\u00BF", "¿")
    txts = txts.replace("\\u00A1", "¡")
    
    txts = txts.replace("\\\\r", "")
    txts = txts.replace("\\\\n", "")
    return txts
```

```python
prices = []
bathrooms = []
rooms = []
surfaces = []
dates = []
neighborhoods = []
zipcodes = []
municipalities = []
agencies = []
latitude = []
longitude = []
descriptions = []
building_type = []
subbuilding_type = []
districts = []
links = []
pools = []
air_con=[]
conserv_state = []
terraces = []
ad_type = []
island = []
community = []
province = []
city = []
building_sub = []
build_type = []
is_new = []
```

```python
aacc = ["asturias", "galicia", "baleares", "cataluna", "extremadura", "region-de-murcia", "andalucia", "aragon", "cantabria", "castilla-la-mancha", "castilla-y-leon", "comunitat-valenciana", "la-rioja", "madrid", "navarra", "pais-vasco",""]
```

```python
for com in aacc:
    for j in range(1,301):
        try:
            with open("./espana/" + com + "pag" + str(j) + ".txt", "r", encoding="utf8") as f:
                page = f.read()

            ad_list = page.split('{\\"accuracy\\"')

            for i in range(1,len(ad_list)):

                #price getter
                try:
                    price = ad_list[i].split('\\"rawPrice\\":')[1]
                    price = price.split(",")[0]
                    prices.append((int(price)))
                except:
                    prices.append(0)

                #bathroom getter
                bathroom = ad_list[i].split('"bathrooms\\",\\"value\\":')
                try:
                    bathroom = bathroom[1].split(",")[0]
                    bathrooms.append(int(bathroom))
                except:
                    bathrooms.append(0)

                #rooms getter
                room = ad_list[i].split('{\\"key\\":\\"rooms\\",\\"value\\":')
                try:
                    room = room[1].split(",")[0]
                    rooms.append(int(room))
                except:
                    rooms.append(0)

                #surface getter
                surface = ad_list[i].split('{\\"key\\":\\"surface\\",\\"value\\":')
                try:
                    surface = surface[1].split(",")[0]
                    surfaces.append(float(surface))
                except:
                    surfaces.append(0)

                #date published
                try:
                    date = ad_list[i].split('\\"timestamp\\":')
                    date = date[1].split(",")[0]
                    date = date.replace("}", "")
                    date = date[:10]
                    dates.append(datetime.fromtimestamp(int(date)))
                except:
                    dates.append(0)

                #neighborhood getter
                try:
                    neighborhood = ad_list[i].split('\\"neighborhood\\":\\"')
                    neighborhood = neighborhood[1].split(",")[0]
                    neighborhood = neighborhood.replace('\\"',"")
                    neighborhoods.append(text_cleaner(neighborhood))
                except:
                    neighborhoods.append("")

                #zipcode getter
                try:
                    zipcode = ad_list[i].split('\\"zipCode\\":\\"')
                    zipcode = zipcode[1].split(",")[0]
                    zipcode = zipcode.replace('\\"',"")
                    zipcodes.append(zipcode)
                except:
                    zipcodes.append(0)

                #municipality getter
                try:
                    municipality = ad_list[i].split('\\"municipality\\":\\"')
                    municipality = municipality[1].split(",")[0]
                    municipality = municipality.replace('\\"',"")
                    municipalities.append(text_cleaner(municipality))
                except:
                    municipalities.append("")

                #real estate agency
                try:
                    agency = ad_list[i].split('\\"clientAlias\\":\\"')
                    agency = agency[1].split(",")[0]
                    agency = agency.replace('\\"',"")
                    agencies.append(agency)
                except:
                    agencies.append("")

                #coordinates getter
                try:
                    coordinates = ad_list[i].split('\\"coordinates\\":{\\"')
                    coordinates = coordinates[1].split(',\\"accuracy\\"')[0]
                    coordinates = coordinates.split(",")
                    lat = coordinates[0]
                    lat = lat.replace('latitude\\":', "")
                    lon = coordinates[1]
                    lon = lon.replace('\\"longitude\\":', "")
                    latitude.append(float(lat))
                    longitude.append(float(lon))
                except:
                    latitude.append(0)
                    longitude.append(0)

                #Description getter
                try:
                    txt = ad_list[i].split('\\"description\\":\\')
                    txt = txt[1].split(',\\"accuracy\\"')[0]
                    txt = txt.split('\\"detail\\')[0]
                    descriptions.append(text_cleaner(txt))
                except:
                    descriptions.append("")

                #Building type getter
                try:
                    build_type = ad_list[i].split('\\"buildingType\\":\\"')
                    build_type = build_type[1].split("\\")
                    building_type.append(build_type[0])
                except:
                    building_type.append("")

                #Building subtype getter
                try:
                    subbuild = ad_list[i].split('"buildingSubtype\\":\\"')
                    subbuild = subbuild[1].split("\\")
                    subbuilding_type.append(subbuild[0])
                except:
                    subbuilding_type.append("")

                #District getter
                try:
                    ditrict = ad_list[i].split('"district\\":\\"')
                    ditrict = ditrict[1].split("\\")
                    districts.append(ditrict[0])
                except:
                    dsitricts.append("")

                #Link getter
                try:
                    x = ad_list[i].split('detail\\":{\\"es-ES\\":\\"')
                    x = x[1].split('\\"},\\"externalContactUrl\\')[0]
                    link = "https://www.fotocasa.es" + x
                    links.append(link)
                except:
                    links.append("")

                #Swimming pool getter
                try:
                    pool = ad_list[i].split('\\"key\\":\\"swimming_pool\\",\\"value\\":')
                    pool = pool[1].split(",\\")
                    pool = pool[0]
                    pools.append(int(pool))
                except:
                    pools.append(0)
                #Air Conditioner getter
                try:
                    ac = ad_list[i].split('"air_conditioner\\",\\"value\\":')
                    ac = ac[1]
                    ac = ac.split(",\\")[0]
                    air_con.append(int(ac))
                except:
                    air_con.append(0)
                #Conservation getter
                try:
                    conserv = ad_list[i].split('"conservationState\\",\\"value\\":')
                    conserv = conserv[1].split(",\\")[0]
                    conserv_state.append(int(conserv))
                except:
                    conserv_state.append(0)
                #Terrace getter
                try:
                    terrace = ad_list[3].split('terrace\\",\\"value\\":')
                    terrace = terrace[1].split(",\\")[0]
                    terraces.append(int(terrace))
                except:
                    terraces.append(0)
                    
                #Community getter
                if com != "":
                    community.append(com)
                elif com == "":
                    community.append("canarias")
                    
                #Province getter
                try:
                    prov = text_cleaner(ad_list[i].split('province\\":\\"')[1].split(",")[0].replace('"', ""))
                    province.append(prov)
                except:
                    province.append("")
                    
                #City getter
                try:
                    cit = text_cleaner(ad_list[i].split(",")[8].split('":\\"')[1])
                    city.append(cit)
                except:
                    city.append("")
                    
                #building subtype
                try:
                    sub = ad_list[i].split('\\"buildingSubtype\\":\\"')[1].split('\\",')[0]
                    building_sub.append(sub)
                except:
                    building_sub.append("")
                    
                #building type
                try:
                    typ = ad_list[i].split('\\"buildingType\\":\\"')[1].split('\\",')[0]
                    build_type.append(typ)
                except:
                    build_type.append("")
                    
                #is new construction
                try:
                    isnew = ad_list[i].split('isNewConstruction\\":')[1].split(',')[0]
                    is_new.append(isnew)
                except:
                    is_new("")
                
        except:
            None
            
```

```python
province_fixed = []
for prov in province:
    province_fixed.append(prov[:len(prov)-1])
```

```python
propiedades = pd.DataFrame()
propiedades['Price'] = pd.Series(prices)
#propiedades['Ad_Type'] = ad_type
propiedades['Bathrooms'] = bathrooms
propiedades['Rooms'] = rooms
propiedades['Surface'] = pd.Series(surfaces)
propiedades['Pools'] = pools
propiedades['Air Conditioner'] = air_con
#propiedades['Type'] = build_type
propiedades['Type'] = building_sub
propiedades['IsNewConstruction'] = is_new
propiedades['Conservation State'] = conserv_state
propiedades['Terrace'] = terraces
#propiedades['Island'] = island
propiedades['Autonomous Community'] = community
propiedades['Province'] = province_fixed
propiedades['City'] = city
propiedades['Neighborhood'] = pd.Series(neighborhoods)
propiedades['Ditricts'] = pd.Series(districts)
propiedades['Municipality'] = pd.Series(municipalities)
propiedades['ZipCode'] = pd.Series(zipcodes)
propiedades['Agency'] = pd.Series(agencies)
propiedades['Latitude'] = pd.Series(latitude)
propiedades['Longitude'] = pd.Series(longitude)
propiedades['Publ_Date'] = pd.Series(dates)
propiedades['Descriptions'] = pd.Series(descriptions)
propiedades['Links'] = links
```

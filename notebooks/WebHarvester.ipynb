```python
import requests
import socks
import socket
import time
import random
```

```python
def getter(url, i, headers, tiempo, start):
    page = requests.get(url + str(i), headers=headers).text
    print("Request done")
    with open ("./espana/pag" + str(i) + ".txt", "w") as f:
        f.write(page)
    print("Archivo " + str(i) + " guardado..." ) 
    end = time.time()
    print("Time elapsed: " + str(((end-start))))
```

```python
start = time.time()
tiempo = [7,8,9,10,11,12,13]
ipprev = 0
count = 0
url = "https://www.fotocasa.es/es/comprar/viviendas/espana/todas-las-zonas/l/"
for i in range(523,22368):
    socks.setdefaultproxy(proxy_type=socks.PROXY_TYPE_SOCKS5, addr="127.0.0.1", port=9050)
    socket.socket = socks.socksocket
    headers = {
        'User-Agent': "Mozilla/5.0 (Windows 98; Win 9x 4.90) AppleWebKit/5332 (KHTML, like Gecko) Chrome/39.0.887.0 Mobile Safari/5332",
        'Accept-Language': "en-US,en;q=0.5",
        'Referer': 'https://google.com',
        'DNT': '1',
    }
    ip = requests.get("https://httpbin.org/ip").text
    ip = ip.split('"')[3]
    print("ip = " + ip)
    if ip == ipprev:
        count+=1
        print("count = " + str(count))
        
        if count < 10:
            getter(url, i, headers, tiempo, start)
            time.sleep(random.choice(tiempo))
            print(" ")
            
        elif count == 10:
            print("Count máximo alcanzado, entrando en sueño...")
            count = 0
            time.sleep(60)
            print("Count a 0...")
            
            getter(url, i, headers, tiempo, start)
            time.sleep(random.choice(tiempo))
            print(" ")
            
    elif ip != ipprev:
        count = 0
        ipprev = ip
        getter(url, i, headers, tiempo, start)
        time.sleep(random.choice(tiempo))
        print(" ")
```
